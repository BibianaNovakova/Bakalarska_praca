```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

ax.axhline(0, linewidth=1)
ax.axvline(0, linewidth=1)

x, y = 2.4, 1.8
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

ax.plot([0, x], [0, y], linewidth=2)
ax.scatter([x], [y], s=35)

phi = np.linspace(0, theta, 200)
ax.plot(0.7*np.cos(phi), 0.7*np.sin(phi), linewidth=1.8)

ax.text(x + 0.08, y + 0.08, r"$z=x+iy$", fontsize=12)
ax.text(x/2 + 0.05, y/2 + 0.15, r"$|z|$", fontsize=12)
ax.text(0.45, 0.18, r"$\arg z$", fontsize=12)
ax.text(3.05, -0.18, r"$Re\{z\}$", fontsize=12)
ax.text(0.12, 3.15, r"$Im\{z\}$", fontsize=12)

ax.set_xlim(-0.8, 3.5)
ax.set_ylim(-0.8, 3.5)
ax.set_aspect("equal", adjustable="box")
ax.set_title("Komplexná rovina a polárny tvar čísla")
ax.grid(True, alpha=0.25)

fig.tight_layout()

plt.show()
```
