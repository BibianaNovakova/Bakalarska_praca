```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.pi / 6
k = 1.3
f = lambda z: k * np.exp(1j * theta) * z

v1 = 1 + 0j
v2 = np.cos(np.pi/4) + 1j*np.sin(np.pi/4)

gamma1 = lambda t: v1*t + 0.5j*t**2
gamma2 = lambda t: v2*t + 0.2j*t**2

t = np.linspace(-0.7, 0.7, 100)
z1, z2 = gamma1(t), gamma2(t)
w1, w2 = f(z1), f(z2)

v1p = f(v1)
v2p = f(v2)

scale_vec = 0.5

fig, axes = plt.subplots(1, 2, figsize=(10,5))

axes[0].plot(z1.real, z1.imag, 'b', label=r'$\gamma_1(t)$')
axes[0].plot(z2.real, z2.imag, 'r', label=r'$\gamma_2(t)$')

axes[0].quiver(
    0, 0,
    scale_vec * v1.real, scale_vec * v1.imag,
    angles='xy', scale_units='xy', scale=1,
    color='b', width=0.008
)

axes[0].quiver(
    0, 0,
    scale_vec * v2.real, scale_vec * v2.imag,
    angles='xy', scale_units='xy', scale=1,
    color='r', width=0.008
)

axes[0].scatter([0], [0], color='k', s=50)
axes[0].text(0.05, -0.1, r'$z_0$', fontsize=12)

axes[0].set_xlabel(r'Re$\{z\}$', fontsize=14)
axes[0].set_ylabel(r'Im$\{z\}$', fontsize=14)
axes[0].tick_params(axis='both', labelsize=12)
axes[0].set_title(r'Rovina $z$', fontsize=14)

axes[0].axis('equal')
axes[0].legend(fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.5)

axes[1].plot(w1.real, w1.imag, 'b', label=r'$f(\gamma_1(t))$')
axes[1].plot(w2.real, w2.imag, 'r', label=r'$f(\gamma_2(t))$')

axes[1].quiver(
    0, 0,
    scale_vec * v1p.real, scale_vec * v1p.imag,
    angles='xy', scale_units='xy', scale=1,
    color='b', width=0.008
)

axes[1].quiver(
    0, 0,
    scale_vec * v2p.real, scale_vec * v2p.imag,
    angles='xy', scale_units='xy', scale=1,
    color='r', width=0.008
)

axes[1].scatter([0], [0], color='k', s=50)
axes[1].text(0.05, -0.1, r'$w_0$', fontsize=12)

axes[1].set_xlabel(r'Re$\{w\}$', fontsize=14)
axes[1].set_ylabel(r'Im$\{w\}$', fontsize=14)
axes[1].tick_params(axis='both', labelsize=12)
axes[1].set_title(r'Rovina $w$', fontsize=14)

axes[1].axis('equal')
axes[1].legend(fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(z):
    return z**2

def f_prime(z):
    return 2*z

# Kružnica mimo (0,0)
t = np.linspace(0, 2*np.pi, 400)
center = 0.9 + 0.5j
radius = 0.4

z = center + radius * np.exp(1j * t)
w = f(z)

# Vybrané body na kružnici
idx = np.linspace(0, len(t)-1, 8, dtype=int)
z_pts = z[idx]

# Dotykové vektory ku kružnici
tangent = 1j * radius * np.exp(1j * t[idx])

# Obrazy bodov a dotykových vektorov
w_pts = f(z_pts)
tangent_w = f_prime(z_pts) * tangent

# Skrátenie vektorov pre čitateľnosť
scale_vec = 0.35
tangent_plot = scale_vec * tangent
tangent_w_plot = scale_vec * tangent_w

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

for ax in axs:
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.tick_params(axis='both', labelsize=12)

axs[0].set_title(r'Rovina $z$', fontsize=16)
axs[0].set_xlabel(r'Re$\{z\}$', fontsize=16)
axs[0].set_ylabel(r'Im$\{z\}$', fontsize=16)

axs[1].set_title(r'Rovina $w$', fontsize=16)
axs[1].set_xlabel(r'Re$\{w\}$', fontsize=16)
axs[1].set_ylabel(r'Im$\{w\}$', fontsize=16)

# Rovina z
axs[0].plot(z.real, z.imag, linewidth=2, label=r'Kružnica v rovine $z$')
axs[0].quiver(
    z_pts.real, z_pts.imag,
    tangent_plot.real, tangent_plot.imag,
    angles='xy',
    scale_units='xy',
    scale=1,
    width=0.005
)
axs[0].legend(fontsize=12)

# Rovina w
axs[1].plot(w.real, w.imag, linewidth=2, label=r'Obraz $w=z^2$')
axs[1].quiver(
    w_pts.real, w_pts.imag,
    tangent_w_plot.real, tangent_w_plot.imag,
    angles='xy',
    scale_units='xy',
    scale=1,
    width=0.005
)
axs[1].legend(fontsize=12)

axs[0].set_xlim(0.3, 1.5)
axs[0].set_ylim(-0.1, 1.1)

axs[1].set_xlim(-0.9, 1.7)
axs[1].set_ylim(0.0, 1.9)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(z):
    return np.exp(z)

def f_prime(z):
    return np.exp(z)

# Hodnoty v bode 0
f0 = f(0)
fprime0 = f_prime(0)

# Lineárna aproximácia
def L(z):
    return f0 + fprime0 * z

# Malá kružnica okolo 0
t = np.linspace(0, 2*np.pi, 200)
radius = 0.6

z = radius * np.exp(1j * t)

# Presný obraz
w = f(z)

# Lineárna aproximácia obrazu
w_lin = L(z)

# Vybrané body a dotykové vektory
idx = np.linspace(0, len(t)-1, 8, dtype=int)

z_pts = z[idx]

# Dotykové vektory na pôvodnej kružnici
tangent = 1j * radius * np.exp(1j * t[idx])

# Dotykové vektory po presnom zobrazení
tangent_w = f_prime(z_pts) * tangent

# Dotykové vektory po lineárnej aproximácii
tangent_w_lin = fprime0 * tangent


fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for ax in axs:
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.tick_params(axis='both', labelsize=12)


axs[0].plot(
    z.real,
    z.imag,
    color='blue',
    linewidth=2,
    label=r'Pôvodná kružnica $z$'
)

axs[0].quiver(
    z_pts.real,
    z_pts.imag,
    tangent.real,
    tangent.imag,
    scale=20,
    width=0.008,
    color='blue'
)

axs[0].set_title(r'Pôvodná kružnica $z$', fontsize=16)

axs[0].set_xlabel(r'$\mathrm{Re}\{z\}$', fontsize=16)
axs[0].set_ylabel(r'$\mathrm{Im}\{z\}$', fontsize=16)

axs[0].legend(fontsize=12, loc='upper right')


axs[1].plot(
    w.real,
    w.imag,
    color='red',
    linewidth=2,
    label=r'Obraz $w=f(z)=e^z$'
)

axs[1].quiver(
    w.real[idx],
    w.imag[idx],
    tangent_w.real,
    tangent_w.imag,
    scale=20,
    width=0.008,
    color='red'
)

axs[1].plot(
    w_lin.real,
    w_lin.imag,
    color='green',
    linewidth=2,
    linestyle='--',
    label=r'Lineárna aproximácia $f(0)+f^\prime(0)z$'
)

axs[1].quiver(
    w_lin.real[idx],
    w_lin.imag[idx],
    tangent_w_lin.real,
    tangent_w_lin.imag,
    scale=20,
    width=0.008,
    color='green'
)

axs[1].set_title(
    r'Obraz a lineárna aproximácia',
    fontsize=16
)

axs[1].set_xlabel(r'$\mathrm{Re}\{w\}$', fontsize=16)
axs[1].set_ylabel(r'$\mathrm{Im}\{w\}$', fontsize=16)

axs[1].legend(fontsize=12, loc='center left',
              bbox_to_anchor=(1.02, 0.5))


plt.tight_layout()
plt.subplots_adjust(wspace=0.35)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.6

f = lambda z: (z - 1j * np.exp(1j*alpha)) / (z + 1j * np.exp(1j*alpha))

fig, axes = plt.subplots(1, 2, figsize=(12,6))
ax1, ax2 = axes


theta = np.linspace(0, 2*np.pi, 400)

circle_x = np.cos(theta)
circle_y = np.sin(theta)

ax2.plot(circle_x, circle_y, linestyle='-', linewidth=1)

ax2.set_aspect('equal', 'box')

ax2.set_title('Obraz v jednotkovom disku')

ax2.set_xlim(-1.5, 1.5)
ax2.set_ylim(-1.5, 1.5)

ax2.axhline(0, linewidth=0.5, color='gray')
ax2.axvline(0, linewidth=0.5, color='gray')

ax2.set_xlabel(r'$\mathrm{Re}\{w\}$', fontsize=13)
ax2.set_ylabel(r'$\mathrm{Im}\{w\}$', fontsize=13)


# hranica: y = tan(alpha)x
xs = np.linspace(-4, 4, 400)
ys = np.tan(alpha) * xs

ax1.plot(xs, ys, color='black', linewidth=1)

ax1.set_title(
    r'Pôvodná polrovina $H_\alpha$: $y>\tan(\alpha)x$'
)

ax1.set_xlim(-4, 4)
ax1.set_ylim(-3, 5)

ax1.set_aspect('equal', 'box')

ax1.set_xlabel(r'$\mathrm{Re}\{z\}$', fontsize=13)
ax1.set_ylabel(r'$\mathrm{Im}\{z\}$', fontsize=13)

# vyfarbenie polroviny

XX, YY = np.meshgrid(
    np.linspace(-4,4,200),
    np.linspace(-3,5,200)
)

mask = YY > np.tan(alpha)*XX

ax1.contourf(
    XX,
    YY,
    mask,
    levels=[0.5,1.5],
    alpha=0.15
)

# rovnobežky s hranicou

offsets = [0.5, 1.0, 2.0]

for off in offsets:
    ys2 = np.tan(alpha)*xs + off

    ax1.plot(
        xs,
        ys2,
        linestyle='--',
        linewidth=0.8
    )

# náhodné body v H_alpha

rng = np.random.default_rng(1)

points = []

while len(points) < 200:

    x = rng.uniform(-3.5, 3.5)

    y = rng.uniform(
        np.tan(alpha)*x + 0.01,
        4.5
    )

    points.append(x + 1j*y)

points = np.array(points)

# body v pôvodnej oblasti
ax1.scatter(
    points.real,
    points.imag,
    s=8,
    alpha=0.6
)

# obrazy bodov
images = f(points)

ax2.scatter(
    images.real,
    images.imag,
    s=8,
    alpha=0.6
)

# hranica -> jednotková kružnica

t = np.linspace(-3, 3, 80)

boundary_points = t * np.exp(1j*alpha)

images_boundary = f(boundary_points)

ax2.plot(
    images_boundary.real,
    images_boundary.imag,
    linestyle='-',
    linewidth=1,
    color='orange',
    label='obraz hranice'
)

# obrazy rovnobežiek

for off in offsets:

    ys2 = np.tan(alpha)*xs + off

    line = xs + 1j*ys2

    mapped_line = f(line)

    ax2.plot(
        mapped_line.real,
        mapped_line.imag,
        linestyle='--',
        linewidth=0.8
    )


ax2.legend()

ax1.grid(True, linestyle=':', alpha=0.4)
ax2.grid(True, linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 30)
y = np.linspace(0, np.pi, 20)

X, Y = np.meshgrid(x, y)
Z_horiz = X + 1j*Y

Xv, Yv = np.meshgrid(x, y)
Z_vert = Xv.T + 1j*Yv.T  # transpozícia, aby boli vertikálne

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
for row in Z_horiz:
    plt.plot(row.real, row.imag, 'b')
for col in Z_vert:
    plt.plot(col.real, col.imag, 'r')
plt.title(r"Pôvodná oblasť A: 0 ≤ $Im\{z\}$ ≤ π")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis('equal')

plt.subplot(1,2,2)
for row in Z_horiz:
    W = np.exp(row)
    plt.plot(W.real, W.imag, 'b')
for col in Z_vert:
    W = np.exp(col)
    plt.plot(W.real, W.imag, 'r')
plt.title(r"Obraz pomocou $f(z)=e^z$")
plt.xlabel(r"Re$\{w\}$")
plt.ylabel(r"Im$\{w\}$")
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inverzná transformácia
def T0_inv(w):
    return 1j * (1 + w) / (1 - w)

r = np.linspace(0.05, 0.95, 20)
theta = np.linspace(0.01, np.pi - 0.01, 30)

R, Theta = np.meshgrid(r, theta)
W = R * np.exp(1j * Theta)

radial_lines = W.T
circular_arcs = W

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Horná polovica jednotkového disku
for row in circular_arcs:
    axs[0].plot(row.real, row.imag, 'b')

for col in radial_lines:
    axs[0].plot(col.real, col.imag, 'r')

# Hranica: horná jednotková polkružnica
t = np.linspace(0.01, np.pi - 0.01, 500)
boundary_w = np.exp(1j * t)

axs[0].plot(boundary_w.real, boundary_w.imag, 'k', linewidth=2,
            label=r"$|w|=1,\ Im\{w\}>0$")

axs[0].set_title("Horná polovica jednotkového disku B")
axs[0].set_xlabel(r"$Re\{w\}$")
axs[0].set_ylabel(r"$Im\{w\}$")
axs[0].grid(True)
axs[0].set_aspect('equal', adjustable='box')
axs[0].set_xlim(-1.2, 1.2)
axs[0].set_ylim(-0.2, 1.2)
axs[0].legend()

# Obraz pomocou T0^{-1}
for row in circular_arcs:
    Z = T0_inv(row)
    axs[1].plot(Z.real, Z.imag, 'b')

for col in radial_lines:
    Z = T0_inv(col)
    axs[1].plot(Z.real, Z.imag, 'r')

# Obraz hornej jednotkovej polkružnice
boundary_z = T0_inv(boundary_w)

mask = np.abs(boundary_z.real) < 10

axs[1].plot(boundary_z.real[mask], boundary_z.imag[mask], 'k',
            linewidth=2, label=r"obraz $|w|=1$")

axs[1].axhline(0, color='gray', linewidth=1)
axs[1].axvline(0, color='gray', linewidth=1)

axs[1].set_title(r"Obraz pomocou $T_0^{-1}(w)$")
axs[1].set_xlabel(r"$Re\{z\}$")
axs[1].set_ylabel(r"$Im\{z\}$")
axs[1].grid(True)
axs[1].set_aspect('equal', adjustable='box')
axs[1].set_xlim(-10, 1)
axs[1].set_ylim(-0.5, 8)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def T0_inv(w):
    return 1j*(1+w)/(1-w)

def final_map(w):
    return ((1+w)/(1-w))**2

theta = np.linspace(0, np.pi, 400)
r = np.linspace(0.05, 0.95, 6)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Horná polovica jednotkového disku
for radius in r:
    w = radius * np.exp(1j * theta)
    axes[0].plot(w.real, w.imag)

for ang in np.linspace(0.1, np.pi - 0.1, 6):
    w = np.linspace(0, 1, 300) * np.exp(1j * ang)
    axes[0].plot(w.real, w.imag)

axes[0].set_title("Horná polovica disku B")
axes[0].set_xlabel(r"$Re\{w\}$")
axes[0].set_ylabel(r"$Im\{w\}$")
axes[0].axhline(0, color="black", linewidth=0.8)
axes[0].axvline(0, color="black", linewidth=0.8)
axes[0].set_aspect("equal")
axes[0].set_xlim(-1.2, 1.2)
axes[0].set_ylim(-0.2, 1.2)
axes[0].grid(True, alpha=0.3)

# Prvý kvadrant
for radius in r:
    w = radius * np.exp(1j * theta)
    z = -1j * T0_inv(w)
    axes[1].plot(z.real, z.imag)

for ang in np.linspace(0.1, np.pi - 0.1, 6):
    w = np.linspace(0, 1, 300) * np.exp(1j * ang)
    z = -1j * T0_inv(w)
    axes[1].plot(z.real, z.imag)

axes[1].set_title("Prvý kvadrant")
axes[1].set_xlabel(r"$Re\{z\}$")
axes[1].set_ylabel(r"$Im\{z\}$")
axes[1].axhline(0, color="black", linewidth=0.8)
axes[1].axvline(0, color="black", linewidth=0.8)
axes[1].set_aspect("equal")
axes[1].set_xlim(-0.5, 3)
axes[1].set_ylim(-0.5, 3)
axes[1].grid(True, alpha=0.3)

# Horná polrovina
for radius in r:
    w = radius * np.exp(1j * theta)
    zeta = final_map(w)
    axes[2].plot(zeta.real, zeta.imag)

for ang in np.linspace(0.1, np.pi - 0.1, 6):
    w = np.linspace(0, 1, 300) * np.exp(1j * ang)
    zeta = final_map(w)
    axes[2].plot(zeta.real, zeta.imag)

axes[2].set_title("Horná polrovina H")
axes[2].set_xlabel(r"$Re\{z\}$")
axes[2].set_ylabel(r"$Im\{z\}$")
axes[2].axhline(0, color="black", linewidth=0.8)
axes[2].axvline(0, color="black", linewidth=0.8)
axes[2].set_aspect("equal")
axes[2].set_xlim(-5, 5)
axes[2].set_ylim(-1, 5)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, -0.01, 400)
y = np.linspace(0, np.pi, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Prvé zobrazenie: exponenciála
W1 = np.exp(Z)

# Druhé zobrazenie: Möbius + druhá mocnina
W2 = ((1 + W1)/(1 - W1))**2

fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(131)

for y0 in np.linspace(0, np.pi, 8):
    ax1.plot(x, y0*np.ones_like(x), color='blue', linewidth=0.8)
for x0 in np.linspace(-4, 0, 8):
    ax1.plot(x0*np.ones_like(y), y, color='red', linewidth=0.8)

# hranice oblasti
ax1.plot(x, np.zeros_like(x), 'k', linewidth=2)        # dolná hranica
ax1.plot(x, np.pi*np.ones_like(x), 'k', linewidth=2)   # horná hranica
ax1.plot(np.zeros_like(y), y, 'k', linewidth=2)        # pravá hranica

ax1.set_title("Oblasť A")
ax1.set_xlim(-4, 4)
ax1.set_ylim(-np.pi, np.pi)
ax1.set_aspect('equal')
ax1.set_xlabel(r"$Re \{z\}$")
ax1.set_ylabel(r"$Im \{z\}$")
ax1.grid(True, linestyle='--', alpha=0.3)

ax2 = fig.add_subplot(132)
for y0 in np.linspace(0.1, np.pi-0.1, 6):
    curve = np.exp(x + 1j*y0)
    ax2.plot(curve.real, curve.imag, 'b', linewidth=0.8)
for x0 in np.linspace(-4, -0.2, 6):
    curve = np.exp(x0 + 1j*y)
    ax2.plot(curve.real, curve.imag, 'r', linewidth=0.8)

circle = np.exp(1j*np.linspace(0, np.pi, 400))
ax2.plot(circle.real, circle.imag, 'k', linewidth=2)

ax2.set_title("Horný polkruh")
ax2.set_aspect('equal')
ax2.set_xlim(-1.2,1.2)
ax2.set_ylim(0,1.2)
ax2.set_xlabel(r"$Re \{w\}$")
ax2.set_ylabel(r"$Im \{w\}$")
ax2.grid(True, linestyle='--', alpha=0.3)

ax3 = fig.add_subplot(133)
for y0 in np.linspace(0.1, np.pi-0.1, 6):
    curve = ((1 + np.exp(x + 1j*y0))/(1 - np.exp(x + 1j*y0)))**2
    ax3.plot(curve.real, curve.imag, 'b', linewidth=0.8)
for x0 in np.linspace(-4, -0.2, 6):
    curve = ((1 + np.exp(x0 + 1j*y))/(1 - np.exp(x0 + 1j*y)))**2
    ax3.plot(curve.real, curve.imag, 'r', linewidth=0.8)

ax3.axhline(0, color='k', linewidth=2)
ax3.set_title("Horná polrovina")
ax3.set_xlim(-10,10)
ax3.set_ylim(0,10)
ax3.set_xlabel(r"$Re \{w\}$")
ax3.set_ylabel(r"$Im \{w\}$")
ax3.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(z):
    return z + 1/z

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)

for theta in np.linspace(0.1, np.pi - 0.1, 10):
    r = np.linspace(1.01, 3, 400)
    z = r * np.exp(1j * theta)
    plt.plot(z.real, z.imag, 'b', linewidth=0.8)

for r in np.linspace(1.1, 3, 8):
    theta = np.linspace(0.01, np.pi - 0.01, 400)
    z = r * np.exp(1j * theta)
    plt.plot(z.real, z.imag, 'r', linewidth=0.8)

theta = np.linspace(0, np.pi, 400)
z = np.exp(1j * theta)
plt.plot(z.real, z.imag, 'k', linewidth=2)

plt.axhline(0, color='k')
plt.axvline(0, color='k', linewidth=0.8)
plt.title(r"Oblasť: $Im\{z\}>0,\ |z|>1$")
plt.xlabel(r"$Re\{z\}$")
plt.ylabel(r"$Im\{z\}$")
plt.gca().set_aspect('equal', 'box')
plt.xlim(-3, 3)
plt.ylim(0, 3)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)

for theta in np.linspace(0.1, np.pi - 0.1, 10):
    r = np.linspace(1.01, 3, 400)
    z = r * np.exp(1j * theta)
    w = f(z)
    plt.plot(w.real, w.imag, 'b', linewidth=0.8)

for r in np.linspace(1.1, 3, 8):
    theta = np.linspace(0.01, np.pi - 0.01, 400)
    z = r * np.exp(1j * theta)
    w = f(z)
    plt.plot(w.real, w.imag, 'r', linewidth=0.8)

plt.axhline(0, color='k')
plt.axvline(0, color='k', linewidth=0.8)
plt.title(r"Obraz: horná polrovina")
plt.xlabel(r"$Re\{w\}$")
plt.ylabel(r"$Im\{w\}$")
plt.gca().set_aspect('equal', 'box')
plt.xlim(-3, 3)
plt.ylim(0, 3)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
